```python
#import libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")

#See the data that has been imported
#df.head(10)
```

```python
#How many records and columns
#df.shape
```

```python
#How everything is associated with eachother
#df.dtypes
```

```python
#Clean up data
#Reduces number of records - Only want USA Races, 50km/50mi, 2020
#Show 50km or 50mi

#df[df['Event distance/length'] == '50km']
```

```python
#Combine 50km/50mi with isin
#Factor only 2020 Races

#df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020)]
```

```python
#Factor only USA Races and clean up
#df[df['Event name'] == 'Everglades 50 Mile Ultra Run (USA)']['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)
```

```python
#Shorten the USA filter
#df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
#Combine all filters together
df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head(10)
```

```python
df2.shape
```

```python
#Remove (USA) from event name
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
#Clean Up athlete age

df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
#Remove h from athlete performance

df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head()
```

```python
#Drop Columns: Athlete club, Athlete country, Athlete year of birth, Athlete age aategory

df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1, inplace = True, errors = 'ignore')
```

```python
df2.head()
```

```python
#Clean up Null Values

df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna()==1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
#Check for Dupes

df2[df2.duplicated() == True]
```

```python
#Reset index

df2.reset_index(drop = True)
```

```python
#Fix types 
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
df2.head()
```

```python
#Rename Columns

#Year of event                  int64
#Event dates                   object
#Event name                    object
#Event distance/length         object
#Event number of finishers      int64
#Athlete performance           object
#Athlete gender                object
#Athlete average speed        float64
#Athlete ID                     int64
#athlete_age                    int32

df2 = df2.rename(columns = {
    'Year of event': 'year',
    'Event dates': 'race_day',
    'Event name':'race_name',
    'Event distance/length': 'race_length',
    'Event number of finishers': 'race_number_of_finishers',
    'Athlete performance': 'athlete_performance',
    'Athlete gender': 'athlete_gender',
    'Athlete average speed': 'athlete_average_speed',
    'Athlete ID': 'athlete_id'    
})
```

```python
df2.head()
```

```python
#Reorder columns

df3 = df2[['race_day', 'race_name', 'race_length', 'race_number_of_finishers', 'athlete_id', 'athlete_gender', 
           'athlete_age', 'athlete_performance', 'athlete_average_speed']]
```

```python
#DATA VISUALISATION
```

```python
#What is the average spped by Race Length?
plt.figure(figsize=(10,6))
sns.boxplot(x='race_length', y='athlete_average_speed', data=df2)
plt.title('Average Speed by Race Length')
plt.xlabel('Race Length')
plt.ylabel('Average Speed')
plt.show()
```

```python
#What is the distribution of genders and age?
plt.figure(figsize=(10,6))
sns.histplot(data=df2, x='athlete_age', hue='athlete_gender', bins=20, kde=True, palette=['blue','pink'], alpha=0.5)
plt.title('Distribution of Athlete Ages by Gender')
plt.xlabel('Age')
plt.ylabel('Number of Athletes')
plt.show()
```

```python
#Correlation Heatmap analysis between the athletes' ages, average speed and number of finshers
sns.heatmap(df2[['athlete_age','athlete_average_speed','race_number_of_finishers']].corr(), annot=True)
```

```python
#Age vs Average Speed
plt.figure(figsize=(10,6))
sns.scatterplot(x='athlete_age', y='athlete_average_speed', hue='athlete_gender', data=df2)
plt.title('Athlete Age vs Average Speed')
plt.xlabel('Age')
plt.ylabel('Average Speed')
plt.show()
```

```python
#Are Younger Athletes quicker than Older atheletes?
corr = df2[['athlete_age', 'athlete_average_speed']].corr()
print("Correlation between age and average speed:")
print(corr)

plt.figure(figsize=(10,6))
sns.regplot(x='athlete_age', y='athlete_average_speed', data=df2, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Age vs Average Speed')
plt.xlabel('Athlete Age')
plt.ylabel('Average Speed')
plt.show()

age_speed = df2.groupby('athlete_age')['athlete_average_speed'].mean()
plt.figure(figsize=(10,6))
age_speed.plot(kind='line', marker='o')
plt.title('Average Speed by Age')
plt.xlabel('Age')
plt.ylabel('Mean Average Speed')
plt.show()
```
